## EDA

```python
print("Average Literacy Rate: ", round(np.mean(df_state_details['literacy_rate']),2),"%")
print("Average Male Literacy Rate: ", round(np.mean(df_state_details['male_literacy_rate']),2),"%")
print("Average Female Literacy Rate: ", round(np.mean(df_state_details['female_literacy_rate']),2),"%")
```

```python
print("State with the highest literacy rate : ",df_state_details[df_state_details['literacy_rate']==np.max(df_state_details['literacy_rate'])].state_name.values[0])
print("State with the lowest literacy rate : ",df_state_details[df_state_details['literacy_rate']==np.min(df_state_details['literacy_rate'])].state_name.values[0])
```

```python
print("State ID with the highest literacy rate : ",df_state_details[df_state_details['literacy_rate']==np.max(df_state_details['literacy_rate'])].state_id.values[0])
print("State ID with the lowest literacy rate : ",df_state_details[df_state_details['literacy_rate']==np.min(df_state_details['literacy_rate'])].state_id.values[0])
```

Factors affecting the Literacy Level (Comparing between the states KERELA and BIHAR):

Family Parameters
- Becoming Head of the family at an early age (employment at an early age)

School Parameters
+ number of schools
+ school in urban or rural area
+ facilities in school
