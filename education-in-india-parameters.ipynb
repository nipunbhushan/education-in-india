```python
import pandas as pd
import numpy as np
```

#### Transforming the Employment Status Information

```python
df_employment=pd.read_csv('download_files/employment_status_gender_age_census_2011.csv')
df_employment.head()

df_employment_status_male=df_employment[['state_id',
                                         'district_id',
                                         'area_name',
                                         'area_type',
                                         'age_group',
                                         'male_main_workers',
                                         'male_non_workers']]

df_employment_status_female=df_employment[['state_id',
                                         'district_id',
                                         'area_name',
                                         'area_type',
                                         'age_group',
                                          'female_main_workers',
                                         'female_non_workers']]

display(df_employment_status_male.head())
display(df_employment_status_female.head())
```

```python
df_employment_status_male_ver= pd.melt(df_employment_status_male,id_vars=['state_id',
                                                                             'district_id',
                                                                             'area_name',
                                                                             'area_type',
                                                                             'age_group'])

df_employment_status_male_ver = df_employment_status_male_ver.rename(columns={'variable':'employment_status','value':'male'})

df_employment_status_male_ver['employment_status']=np.where(df_employment_status_male_ver['employment_status']=='male_main_workers','employed','unemployed')

df_employment_status_male_ver.head()
```

```python
df_employment_status_female_ver= pd.melt(df_employment_status_female,id_vars=['state_id',
                                                                             'district_id',
                                                                             'area_name',
                                                                             'area_type',
                                                                             'age_group'])

df_employment_status_female_ver = df_employment_status_female_ver.rename(columns={'variable':'employment_status','value':'female'})

df_employment_status_female_ver['employment_status']=np.where(df_employment_status_female_ver['employment_status']=='female_main_workers','employed','unemployed')

df_employment_status_female_ver.head()
```

```python
df_employment_status = pd.merge(left = df_employment_status_male_ver,
                                 right = df_employment_status_female_ver,
                                 how = 'inner', 
                                 left_on = ['state_id',
                                             'district_id',
                                             'area_name',
                                             'area_type',
                                             'age_group',
                                            'employment_status'], 
                                 right_on= ['state_id',
                                             'district_id',
                                             'area_name',
                                             'area_type',
                                             'age_group',
                                            'employment_status'])

df_employment_status.head()
```

```python
df_employment_status=df_employment_status.drop(['area_name','district_id'], axis=1)
df_employment_status=df_employment_status.reset_index(drop=True)
display(df_employment_status.head())
df_employment_status.to_csv('transformed_files/cleaned_employment_status.csv')
```

#### Transforming the Marital Status Information

```python
df_marital_status_src=pd.read_csv('download_files/household_head_marital_gender_age_census_india_2011.csv')
df_marital_status_src.head()
```

```python
df_marital_status_male=df_marital_status_src[['state_id',
                                         'district_id',
                                         'state_name',
                                         'area_type',
                                         'age_group_of_head',
                                         'male_head_household',
                                         'male_head_never_married',
                                        'male_head_currently_married',
                                        'male_head_widowed',
                                        'male_head_divorced',
                                        'male_head_separated']]

df_marital_status_female=df_marital_status_src[['state_id',
                                         'district_id',
                                         'state_name',
                                         'area_type',
                                         'age_group_of_head',
                                         'female_head_household',
                                         'female_head_never_married',
                                        'female_head_currently_married',
                                        'female_head_widowed',
                                        'female_head_divorced',
                                        'female_head_separated']]

display(df_marital_status_male.head())
display(df_marital_status_female.head())
```

```python
df_marital_status_male_ver= pd.melt(df_marital_status_male,id_vars=['state_id',
                                                                     'district_id',
                                                                     'state_name',
                                                                     'area_type',
                                                                     'age_group_of_head'])

df_marital_status_male_ver = df_marital_status_male_ver.rename(columns={'variable':'marital_status','value':'male'})

df_marital_status_male_ver = df_marital_status_male_ver.replace({'marital_status' : {'male_head_household' : 'Total', 
                                                                                           'male_head_never_married' : 'Single', 
                                                                                           'male_head_currently_married' : 'Married', 
                                                                                           'male_head_widowed' : 'Widowed', 
                                                                                           'male_head_divorced' : 'Divorced', 
                                                                                           'male_head_separated' : 'Separated'}})

df_marital_status_male_ver.head()
```

```python
df_marital_status_female_ver= pd.melt(df_marital_status_female,id_vars=['state_id',
                                                                         'district_id',
                                                                         'state_name',
                                                                         'area_type',
                                                                         'age_group_of_head'])

df_marital_status_female_ver = df_marital_status_female_ver.rename(columns={'variable':'marital_status','value':'female'})

df_marital_status_female_ver = df_marital_status_female_ver.replace({'marital_status' : {'female_head_household' : 'Total', 
                                                                                           'female_head_never_married' : 'Single', 
                                                                                           'female_head_currently_married' : 'Married', 
                                                                                           'female_head_widowed' : 'Widowed', 
                                                                                           'female_head_divorced' : 'Divorced', 
                                                                                           'female_head_separated' : 'Separated'}})

df_marital_status_female_ver.head()
```

```python
df_marital_status = pd.merge(left = df_marital_status_male_ver,
                                 right = df_marital_status_female_ver,
                                 how = 'inner', 
                                 left_on = ['state_id',
                                             'district_id',
                                             'state_name',
                                             'area_type',
                                             'age_group_of_head',
                                            'marital_status'], 
                                 right_on= ['state_id',
                                             'district_id',
                                             'state_name',
                                             'area_type',
                                             'age_group_of_head',
                                            'marital_status'])

df_marital_status.head()
```

```python
df_marital_status=df_marital_status.drop(['district_id','state_name'], axis=1)
df_marital_status=df_marital_status.reset_index(drop=True)
display(df_marital_status.head())
df_marital_status.to_csv('transformed_files/cleaned_marital_status.csv')
```
