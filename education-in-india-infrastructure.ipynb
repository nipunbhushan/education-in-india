```python
import pandas as pd
import numpy as np
df_state_elem_meta=pd.read_csv("data_files/2015_16_Statewise_Elementary_Metadata.csv")
df_state_elem=pd.read_csv("data_files/2015_16_Statewise_Elementary.csv")
df_state_elem=df_state_elem.drop(['AC_YEAR'],axis=1)
df_state_elem_meta=df_state_elem_meta.iloc[1:].reset_index(drop=True)
```

#### Transforming the State Information into State Details DataFrame

```python
df_state_details=df_state_elem.iloc[:,:19]
df_state_details = df_state_details.drop(['DISTRICTS','BLOCKS','CLUSTERS','TOTPOPULAT','P_URB_POP','GROWTHRATE','SEXRATIO','OVERALL_LI','AREA_SQKM'], axis=1)
df_state_details = df_state_details.rename(columns={'STATCD' : 'state_id',
                                                    'STATNAME' : 'state_name',
                                                    'VILLAGES' : 'villages',
                                                    'POPULATION_0_6' : 'population_0_6',
                                                    'P_SC_POP' : 'perc_SC_population',
                                                    'P_ST_POP' : 'perc_ST_population',
                                                    'FEMALE_LIT' : 'female_literacy_rate',
                                                    'MALE_LIT' : 'male_literacy_rate',
                                                    'TOT_6_10_15' : 'population_6_10',
                                                    'TOT_11_13_15' : 'population_11_13'})
df_imported_census=pd.read_csv("data_files/imported_census_2011.csv")
df_state_details = pd.merge(left = df_state_details,
                                 right = df_imported_census,
                                 how = 'inner', 
                                 left_on = "state_id", 
                                 right_on= "state_id")
df_state_details['SC_population']=df_state_details['population']*df_state_details['perc_SC_population']
df_state_details['ST_population']=df_state_details['population']*df_state_details['perc_ST_population']
df_state_details = df_state_details.drop(['perc_SC_population','perc_ST_population'], axis=1)
df_state_details=df_state_details[['state_id',
                                   'state_name',
                                   'villages',
                                   'population',
                                   'perc_to_total_population',
                                   'males',
                                   'females',
                                   'sex_ratio',
                                   'population_0_6',
                                   'population_6_10',
                                   'population_11_13',
                                   'rural_population',
                                   'urban_population',
                                   'SC_population',
                                   'ST_population',
                                   'area_sqkm',
                                   'density_per_sqkm',
                                   'growth_rate',
                                   'male_literacy_rate',
                                   'female_literacy_rate',
                                   'literacy_rate']]

df_state_details=df_state_details[df_state_details['state_id']!=36]

display(df_state_details.shape)
display(df_state_details.head())
df_state_details.to_csv('transformed_files/cleaned_state_details.csv',index=False)
```

#### Transforming the Schools Information into a School DataFrame

```python
df_schools=pd.concat([df_state_elem['STATCD'],df_state_elem['SCHTOT']],axis=1)
df_schools.columns=['state_id','total']
df_schools['government']=df_state_elem['SCHTOTG']
df_schools['private']=df_state_elem['SCHTOTP']
df_schools['madarsas']=df_state_elem['SCHTOTM']
df_schools['rural']=df_state_elem['SCHTOTGR']
df_schools['government_aided']=df_state_elem['SCHTOTGA']
df_schools['private_rural']=df_state_elem['SCHTOTPR']
df_schools['boys_only']=df_state_elem['SCHBOYTOT']
df_schools['girls_only']=df_state_elem['SCHGIRTOT']
df_schools['single_classroom']=df_state_elem['SCLSTOT']
df_schools['single_teacher']=df_state_elem['STCHTOT']
df_schools['road_approachable']=df_state_elem['ROADTOT']
df_schools['playground_facility']=df_state_elem['SPLAYTOT']
df_schools['boundary_wall']=df_state_elem['SBNDRTOT']
df_schools['girls_toilet']=df_state_elem['SGTOILTOT']
df_schools['boys_toilet']=df_state_elem['SBTOILTOT']
df_schools['drinking_water']=df_state_elem['SWATTOT']
df_schools['electricity']=df_state_elem['SELETOT']
df_schools['computers']=df_state_elem['SCOMPTOT']
df_schools['with_ramp']=df_state_elem['SRAMTOT']
df_schools['no_ramp']=df_state_elem['SRAMNTOT']
df_schools['established_2001']=df_state_elem['ESTDTOT']
df_schools['midday_meal']=df_state_elem['MDMTOT']
df_schools['kitchen_shed']=df_state_elem['KITTOT']
df_schools['midday_kitchen']=df_state_elem['KITSTOT']
df_schools['less_than_50_students']=df_state_elem['ENR50TOT']
df_schools['school_mgmt_committee']=df_state_elem['SMCTOT']
df_schools['preprimary']=df_state_elem['PREP']
df_schools['TLM_grant']=df_state_elem['TLM_R1']+df_state_elem['TLM_R2']+df_state_elem['TLM_R3']+df_state_elem['TLM_R4']+df_state_elem['TLM_R5']+df_state_elem['TLM_R6']+df_state_elem['TLM_R7']
df_schools['school_development_grant']=df_state_elem['CONTI_R1']+df_state_elem['CONTI_R2']+df_state_elem['CONTI_R3']+df_state_elem['CONTI_R4']+df_state_elem['CONTI_R5']+df_state_elem['CONTI_R6']+df_state_elem['CONTI_R7']

df_schools=df_schools[df_schools['state_id']!=36]

display(df_schools.shape)
display(df_schools.head())

df_schools_ver= pd.melt(df_schools,id_vars=['state_id'])
df_schools_ver = df_schools_ver.rename(columns={'variable':'school_type','value':'num_of_schools'})

display(df_schools_ver.shape)
display(df_schools_ver.head())

df_schools_ver.to_csv('transformed_files/cleaned_schools.csv',index=False)
```

```python
df_students=pd.concat([df_state_elem['STATCD'],df_state_elem['ENRTOT']],axis=1)
df_students.columns=['state_id','total']
df_students['girls']=df_state_elem['ENRGTOT']
df_students['boys']=df_students['total']-df_students['girls'] #derived
df_students['government']=df_state_elem['ENRTOTG']
df_students['private']=df_state_elem['ENRTOTP']
df_students['madarsas']=df_state_elem['ENRTOTM']
df_students['government_rural']=df_state_elem['ENRTOTGR']
df_students['private_rural']=df_state_elem['ENRTOTPR']
df_students['preprimary']=df_state_elem['PRESTD']
df_students['SC']=df_state_elem['SCPTOT']+df_state_elem['SCUTOT']
df_students['SC_girls']=df_state_elem['SCPTOT_G']+df_state_elem['SCUTOT_G']
df_students['SC_boys']=df_students['SC']-df_students['SC_girls']
df_students['ST']=df_state_elem['STPTOT']+df_state_elem['STUTOT']
df_students['ST_girls']=df_state_elem['STPTOT_G']+df_state_elem['STUTOT_G']
df_students['ST_boys']=df_students['ST']+df_students['ST_girls']
df_students['OBC']=df_state_elem['OBPTOT']+df_state_elem['OBUTOT']
df_students['OBC_girls']=df_state_elem['OBPTOT_G']+df_state_elem['OBUTOT_G']
df_students['OBC_boys']=df_students['OBC']+df_students['OBC_girls']
df_students['Muslims']=df_state_elem['MUPTOT']+df_state_elem['MUUTOT']
df_students['Muslims_girls']=df_state_elem['MUPTOT_G']+df_state_elem['MUUTOT_G']
df_students['Muslims_boys']=df_students['Muslims']+df_students['Muslims_girls']
# Grade 1_9
df_students['repeater_boys']=df_state_elem['C1_BR']+df_state_elem['C2_BR']+df_state_elem['C3_BR']+df_state_elem['C4_BR']+df_state_elem['C5_BR']+df_state_elem['C6_BR']+df_state_elem['C7_BR']+df_state_elem['C8_BR']+df_state_elem['C9_BR']
df_students['repeater_girls']=df_state_elem['C1_GR']+df_state_elem['C2_GR']+df_state_elem['C3_GR']+df_state_elem['C4_GR']+df_state_elem['C5_GR']+df_state_elem['C6_GR']+df_state_elem['C7_GR']+df_state_elem['C8_GR']+df_state_elem['C9_GR']
# Grade 1_8
df_students['disability_boys']=df_state_elem['C1_BD']+df_state_elem['C2_BD']+df_state_elem['C3_BD']+df_state_elem['C4_BD']+df_state_elem['C5_BD']+df_state_elem['C6_BD']+df_state_elem['C7_BD']+df_state_elem['C8_BD']
df_students['disability_girls']=df_state_elem['C1_GD']+df_state_elem['C2_GD']+df_state_elem['C3_GD']+df_state_elem['C4_GD']+df_state_elem['C5_GD']+df_state_elem['C6_GD']+df_state_elem['C7_GD']+df_state_elem['C8_GD']

df_students=df_students[df_students['state_id']!=36]

display(df_students.shape)
display(df_students.head())

df_students_ver= pd.melt(df_students,id_vars=['state_id'])
df_students_ver = df_students_ver.rename(columns={'variable':'student_type','value':'num_of_students'})

display(df_students_ver.shape)
display(df_students_ver.head())


df_students_ver.to_csv('transformed_files/cleaned_students.csv',index=False)
```

#### Transforming the Education Level by Age Group and Literacy Information into a DataFrame

```python
df_education_level_students_male=df_education_level_students[['state_id',
                                                                'state_name',
                                                                'area_type',
                                                                'age_group',
                                                                'total_males',
                                                                'illiterate_males',
                                                                'literate_males',
                                                                'literate_ne_males',
                                                                'preprimary_males',
                                                                'primary_males',
                                                                'middle_males',
                                                                'secondary_males',
                                                                'higher_secondary_males',
                                                                'non_tech_diploma_males',
                                                                'tech_diploma_males',
                                                                'graduate_males']]

df_education_level_students_female=df_education_level_students[['state_id',
                                                                'state_name',
                                                                'area_type',
                                                                'age_group',
                                                                'total_females',
                                                                'illiterate_females',
                                                                'literate_females',
                                                                'literate_ne_females',
                                                                'preprimary_females',
                                                                'primary_females',
                                                                'middle_females',
                                                                'secondary_females',
                                                                'higher_secondary_females',
                                                                'non_tech_diploma_females',
                                                                'tech_diploma_females',
                                                                'graduate_females']]

display(df_education_level_students_male.head())
display(df_education_level_students_female.head())
```

```python
df_education_level_students_male_ver= pd.melt(df_education_level_students_male,id_vars=['state_id',
                                                                    'state_name',
                                                                    'area_type',
                                                                    'age_group'])

df_education_level_students_male_ver = df_education_level_students_male_ver.rename(columns={'variable':'education_level','value':'male'})

df_education_level_students_male_ver = df_education_level_students_male_ver.replace({'education_level' : {'total_males' : 'Total', 
                                                                                                            'illiterate_males' : 'Illiterate', 
                                                                                                            'literate_males' : 'Literate', 
                                                                                                            'literate_ne_males' : 'Literate with No Education', 
                                                                                                            'preprimary_males' : 'Preprimary', 
                                                                                                            'primary_males' : 'Primary',
                                                                                                            'middle_males' : 'Middle',
                                                                                                            'secondary_males' : 'Secondary',
                                                                                                            'higher_secondary_females' : 'Higher Secondary',
                                                                                                            'non_tech_diploma_males' : 'Non-Technical Diploma',
                                                                                                            'tech_diploma_females' : 'Technical Diploma',
                                                                                                            'graduate_males' : 'Graduate'}})

df_education_level_students_male_ver.head()
```

```python
df_education_level_students_female_ver= pd.melt(df_education_level_students_female,id_vars=['state_id',
                                                                    'state_name',
                                                                    'area_type',
                                                                    'age_group'])

df_education_level_students_female_ver = df_education_level_students_female_ver.rename(columns={'variable':'education_level','value':'female'})

df_education_level_students_female_ver = df_education_level_students_female_ver.replace({'education_level' : {'total_females' : 'Total', 
                                                                                                            'illiterate_females' : 'Illiterate', 
                                                                                                            'literate_females' : 'Literate', 
                                                                                                            'literate_ne_females' : 'Literate with No Education', 
                                                                                                            'preprimary_females' : 'Preprimary', 
                                                                                                            'primary_females' : 'Primary',
                                                                                                            'middle_females' : 'Middle',
                                                                                                            'secondary_females' : 'Secondary',
                                                                                                            'higher_secondary_fefemales' : 'Higher Secondary',
                                                                                                            'non_tech_diploma_females' : 'Non-Technical Diploma',
                                                                                                            'tech_diploma_fefemales' : 'Technical Diploma',
                                                                                                            'graduate_females' : 'Graduate'}})

df_education_level_students_female_ver.head()
```

```python
df_education_level['total']=df_education_level['male']+df_education_level['female']
display(df_education_level.head())
df_education_level.to_csv('transformed_files/cleaned_education_level.csv')
```
